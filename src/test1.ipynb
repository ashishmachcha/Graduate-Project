```python
from langgraph.checkpoint.memory import InMemorySaver 
from AI.agents import get_cursor_agent
```

```python
import uuid 

# Config MUST include: user_id, project_id (and optionally run_id for locking)
config = {
    "configurable": {
        "user_id": "1",
        "project_id": "test_project_2",  # REQUIRED: All tools need this
        "run_id": str(uuid.uuid4()),  # Optional but recommended for project locking
        "thread_id": str(uuid.uuid4()),  # For LangGraph checkpointer
    }
}

checkpointer = InMemorySaver()
agent = get_cursor_agent(model="llama-3.1-8b-instant", checkpointer=checkpointer)
```

```python
# Create workspace manually (tools are LangChain StructuredTool objects, not callable directly)
from pathlib import Path
import os
from django.conf import settings

# Get workspace directory (same logic as tools.py uses)
try:
    workspace_base = Path(getattr(settings, "AI_WORKSPACE_DIR", None) or os.getenv("AI_WORKSPACE_DIR", "workspace")).resolve()
except:
    workspace_base = Path(os.getenv("AI_WORKSPACE_DIR", "workspace")).resolve()

workspace_path = workspace_base / "3" / "test_project_1"
workspace_path.mkdir(parents=True, exist_ok=True)
print(f"✅ Workspace created: {workspace_path}")

```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "can you create a django to do list app with front end and back end can you create full project just dont tell me what to do just give me full project"}
        ]
    },
    config
)

response['messages'][-1].content
```

```python
response = agent.invoke(
    {"messages": 
        [
            {"role": "user", "content": "add one more model in todo list app called category with name and description and created at"}
        ]
    },
    config
)

response['messages'][-1].content
```

```python
response = agent.invoke(
    {"messages": 
        [
            {
                "role": "system",
                "content": """You are a coding assistant. You MUST use tools to modify code.
Use only these tools: read_file, write_file, apply_patch, list_files, search_text, run_command.
To change code:
1) First call read_file on the target file.
2) Then call write_file OR apply_patch to write the updated content."""
            },
            {
                "role": "user",
                "content": """In the workspace project, update the backend todo list app models file
(tod o_list_app_backend/models.py or similar) to add a new model Category with fields:
- name (CharField, max_length=100)
- description (TextField, blank=True, null=True)
- created_at (DateTimeField, auto_now_add=True)

Steps:
1) Use read_file to inspect the existing models.py.
2) Then write a new models.py that contains both the existing TodoItem model and the new Category model."""
            }
        ]
    },
    config
)

print(response["messages"][-1].content)
```
